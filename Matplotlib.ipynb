```python
import pandas as pd

batsman = pd.read_csv('matches - matches.csv')
batsman.head()
```

```python
subset = batsman.head(5)

plt.plot(subset['id'], subset['team1'], color='#D9F10F', linestyle='solid', label='team1', marker='o')
plt.plot(subset['id'], subset['team2'], color='#FC00D6', linestyle='dashdot', label='team2', marker='D')

plt.title('Team1 vs Team2 (First 5 Matches)')
plt.xlabel('Match ID')
plt.ylabel('Teams')
plt.legend()
plt.grid()
plt.show()

```

```python
import matplotlib.pyplot as plt

price = [4500, 5400, 6300, 7900, 8500, 9000, 1700]
year = [2015, 2021, 2023, 2025, 2017, 2012, 2024]

# Basic bar plot
plt.bar(year, price, color='skyblue', label='All Prices')

# Highlight specific bars by plotting them again on top
plt.bar(year[0], price[0], color='red', label='4500 in 2015')
plt.bar(year[6], price[6], color='blue', label='1700 in 2024')

plt.title('Bar Chart of Price vs Year')
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend()
plt.grid(axis='y')
plt.show()

```
