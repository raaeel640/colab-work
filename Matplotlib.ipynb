```python
import matplotlib.pyplot as plt

price = [4500, 5400, 6300, 7900, 8500, 9000, 1700]
year = [2015, 2021, 2023, 2025, 2017, 2012, 2024]

# Basic bar plot
plt.bar(year, price, color='skyblue', label='All Prices')

# Highlight specific bars by plotting them again on top
plt.bar(year[0], price[0], color='red', label='4500 in 2015')
plt.bar(year[6], price[6], color='blue', label='1700 in 2024')

plt.title('Bar Chart of Price vs Year')
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend()
plt.grid(axis='y')
plt.show()

```

```python
import pandas as pd

batter = pd.read_csv('batter.csv')
batter.head(25)
```

```python
import matplotlib.pyplot as plt
subset = batter.head(25)

plt.figure(figsize=(15,7))
plt.scatter(subset['avg'], subset['strike_rate'], color='green', marker='o')
plt.title('Average vs Strike Rate (Top 5 Batters)')
plt.xlabel('Batting Average')
plt.ylabel('Strike Rate')
plt.grid()
for i in range(subset.shape[0]):
    plt.text(subset['avg'].values[i],subset['strike_rate'].values[i],subset['batter'].values[i])
plt.show()

```

```python
fig,ax =plt.subplots(figsize=(15,7))
ax.scatter(batter['avg'], batter['strike_rate'],color='red')
ax.set_title("Scatter Plot of Average vs Strike Rate")
ax.set_xlabel("Average")
ax.set_ylabel('Strike Rate')
```

```python
fig ,ax = plt.subplots(nrows=2,ncols=1)

ax[0].scatter(subset['avg'],subset['strike_rate'])
ax[1].scatter(subset['avg'],subset['runs'])
```

```python
import numpy as np
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
(100, 100)
```

```python
z = xx**2 + yy**2
z.shape
(100, 100)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
