```python
def run_scored(batsman_name):
  vk=data[data['batsman']==batsman_name]
  return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
run_scored('Ms Dhoni')
```

```python
#isin
# find the most destructive daeth over batsman in the history of IPL
# strike rate =(number ofruns/number of balls)/100
# min batsman 200 balls in over 16-20
data.groupby('batsman')['batsman_runs'].count()
```

```python
death_over =data[data['over']>15]
all_batsman =death_over.groupby('batsman')['batsman_runs'].count()>200
x=all_batsman>200
batsman_list=all_batsman[x].index.tolist()
```

```python
# runs scored by all these 43 batsmen
# balls played by all these 43 batsmen
final =data[data['batsman'].isin(batsman_list)]
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
runs
```

```python
balls=final.groupby('batsman')['batsman_runs'].count()
```

```python
sr=(runs/balls)*100
sr
```
