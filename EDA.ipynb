2 CLEANING THE DATA


```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df=df.fillna('')
df
```

```python
df.drop(['Gross'],axis=1,inplace=True)
```

```python
df['YEAR'] = df['YEAR'].str.extract(r'(\d{4})')
df
```

```python
# median_value = df['RATING'].median()
# print("Median:", median_value)

mean_value = df['RATING'].mean()
print("Median:", mean_value)
# mean was less than median so we used mean to fill the empty spaces
```

```python
df['RATING'] = pd.to_numeric(df['RATING'])
```

```python
mean_value = df['RATING'].mean()
df['RATING'].fillna(mean_value)

```

```python
df['RunTime'] = pd.to_numeric(df['RunTime'])
```

```python
median_value = df['RunTime'].median()
print("Median:", median_value)
```

```python
mean_value = df['RunTime'].mean()
df['RunTime'].fillna(median_value)
```
